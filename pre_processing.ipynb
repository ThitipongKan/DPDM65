ข้อมูล missing: nul, Nan (not a number), Na (not authorised/ not available)

สามารถเข้าไปเช็คข้อมูลได้โดยพิมพ์เข้า google: dropna pandas --
จะเห็นว่า dropna เป็นฟังก์ชันที่มี input 5 ตัว (มี default ทั้งหมด)

DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
- axis: 0 = ดูตามแถว / 1 = ดูตามคอลัมน์
- how: any = ถ้ามี missing สักอัน จะตัด axis นั้นออกไป / all = ต้องเป็น missing ทั้ง axis ถึงจะตัดออก
- thresh (threshold): กำหนดจำนวน non-NA ถ้ามีน้อยกว่าที่กำหนดจะตัดแถวนั้นออกไป
- subset: พิจารณา missing เฉพาะคอลัมน์ที่กำหนด (ถ้ามีจะตัดแถวนั้นทิ้งไป) - เป็นการตัดโดยทำให้คอลัมน์ที่จะได้ใช้งานต่อมีค่าเสมอ
- inplace: true = อัปเดตตัวแปรเดิมให้เป็นตัวใหม่ที่ทำ (มีค่าเท่ากับการเอาตัวแปรใหม่มารับ) -- ปกติไม่ค่อยใช้ เพราะจะทำให้ไล่ยาก ด้วยความไม่รู้ว่าตัวแปรปัจจุบันถูกเปลี่ยนไปเป็นอะไรตั้งแต่เมื่อไหร่ มักจะเอาตัวแปรอื่นมารับมากกว่า

## สร้างตารางโดยวิธี Dictionary

### Dict-Row (ทีละแถว)

````python
sale = [{'account':'Jones LLC', 'Jan':150, 'Feb':200, 'Mar':140},
           {'account':'Alpha Co', 'Jan':200, 'Feb':210, 'Mar':215},
           {'account':'Blue Inc', 'Jan':50, 'Feb':90, 'Mar':95}
          ]

sale_tb1 = pd.DataFrame(sale)
sale_tb1
````

### Dict-Column (สร้างตาราง ทีละคอลัมน์ !นิยม)

````python
sales = {'account' : ['Jones LLC', 'Alpha Co', 'Blue Inc'],
             'Jan'       : [150, 200, 50],
             'Feb'       : [200, 210, 90],
             'Mar'       : [140,215,95]
            }

sale_tb2 = pd.DataFrame.from_dict(sales)
sale_tb2
````

## สร้างตารางโดยวิธี List

### List-Row

````python
saleL = [('Jones LLC', 150, 200, 140), ('Alpha Co', 200, 210, 215), ('Blue Inc', 50, 90, 95)]
labels = ('account', 'Jan', 'Feb', 'Mar')

sale_table3 = pd.DataFrame.from_records(saleL, columns=labels)
sale_table3
````

สร้างตาราง ที่ประกอบด้วย
- สูตรคูณแม่ 1-5
- factorial
- ยกกำลังสอง
- เลข 3 หารค่าในช่องยกกำลังสองลงตัว (yes/no)

สำหรับเลข 1-15 (เป็นตาราง 8 คอลัมน์, 15 แถว)

````python
def factorial(n):
    return 1 if (n==1 or n==0) else n * factorial(n - 1);
````

````python
for i in list(range(1,16)) :        #ไม่ได้ใช้ต่อ!
    y1 = i*1
    y2 = i*2
    y3 = i*3
    y4 = i*4
    y5 = i*5
    fac = factorial(i)
    sq = i*i
    if ( (sq % 3) == 0 ):
        dv = 'yes'
    else :
        dv = 'no'
    value = [y1, y2, y3, y4, y5, fac, sq, dv]
    print(value)
````

````python
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
fact = 1
fac = []
sq = []
dvs = []

for i in list(range(1,16)) :
    y1 = y1+[i*1]
    y2 = y2+[i*2]
    y3 = y3+[i*3]
    y4 = y4+[i*4]
    y5 = y5+[i*5]
    fact = fact*i
    fac = fac+[fact]
    sqi = i*i
    sq = sq+[sqi]
    if (sqi % 3) == 0 :
        dv = 'yes'
    else :
        dv = 'no'
    dvs = dvs+[dv]
````

````python
Tab = {'x1':y1, 'x2':y2, 'x3':y3, 'x4':y4, 'x5':y5, 'Factorial':fac, 'Square':sq, 'Divisible#3':dvs}

Table = pd.DataFrame.from_dict(Tab)
Table
````
